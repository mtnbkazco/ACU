```python
all_of_the_data = []

for single_month in folder:
    try:
        nested_dict = pd.read_excel(single_month, sheet_name = None)

        list_of_dfs = []
        for i in nested_dict.keys():
            list_of_dfs.append(nested_dict[i])

        date = list_of_dfs[0].columns[0].strftime("%Y-%m-%d")

        list_of_cleaned_df = []
        for df in list_of_dfs:
            df = df.fillna(value=np.nan)

            try:
                class_code = df.iloc[:,1:2].values.tolist()
                class_code = [item for sublist in class_code for item in sublist]
                class_code = list(set(class_code))
                class_code = [i for i in class_code if len(str(i)) > 9][1]
                class_code = class_code.replace("\n", "")
                class_code = class_code.replace("-", "")
                class_code = class_code.replace("  ", " ")
            except:
                pass
            df = df.dropna(thresh=13)
            column_s = df.dropna(thresh=15).values.tolist()[0]
            df.columns = column_s
            df = df[1:]
            df = df.loc[:, df.columns.notnull()]
            try:
                df["Class Code"] = class_code
            except:
                df["Class Code"] = "nan"
            df = df.reset_index(drop = True)

            list_of_cleaned_df.append(df)

        #print(len(list_of_cleaned_df))

        total_month = pd.concat(list_of_cleaned_df).reset_index(drop=True)
        #print(total_month.head())
        total_month["Date"] = date
        total_month = total_month[~total_month["Class Code"].str.startswith('ZZZ')] # drop summary tab
        total_month.columns = ['Class Code', 'Item ID', 'Bottle Size', 'Product Name', 'Status', 'Price', 'Cost', 'Total Sales', 'Percent Sales', 'YOY Sales', 'Total Bottle Count','Percent Bottle Count', 'YOY Bottle Count', 'Case Count (9L)', 'Rank', 'Date'] 

        total_month['Item ID'] = pd.to_numeric(total_month['Item ID'], errors='coerce')
        total_month['Bottle Size'] = pd.to_numeric(total_month['Bottle Size'], errors='coerce')
        total_month['Price'] = pd.to_numeric(total_month['Price'], errors='coerce')
        total_month['Cost'] = pd.to_numeric(total_month['Cost'], errors='coerce')
        total_month['Total Sales'] = pd.to_numeric(total_month['Total Sales'], errors='coerce')
        all_of_the_data.append(total_month)
    except:
        print(date)
```

```python
#April2018
April2018 = pd.read_excel(folder[4], sheet_name = None)

list_of_dfs = []
for i in April2018.keys():
    list_of_dfs.append(April2018[i])

date = list_of_dfs[0].columns[0].strftime("%Y-%m-%d")
date
```

```python
list_of_cleaned_df = []
for df in list_of_dfs:
    df = df.fillna(value=np.nan)

list_of_cleaned_df = []
for df in list_of_dfs:
    df['Unnamed: 1'] = df['Unnamed: 1'].ffill(axis = 0)
    df['Unnamed: 8'] = df['Unnamed: 8'].bfill(axis = 0)
    df['Unnamed: 9'] = df['Unnamed: 9'].bfill(axis = 0)
    df = df[df['Unnamed: 19'].apply(lambda x: str(x).isdigit())]
    df = df[df['Unnamed: 2'].apply(lambda x: str(x).isdigit())]
    df = df.dropna(thresh=11)
    df = df.dropna(how='all', axis=1)
    df.columns = ['Class Code', 'Item ID', 'Bottle Size', 'Product Name', 'Status', 'Price', 'Cost', 'Total Sales', 'Percent Sales', 'YOY Sales', 'Total Bottle Count','Percent Bottle Count', 'YOY Bottle Count', 'Case Count (9L)', 'Rank'] 
    df["Class Code"] = df["Class Code"].str.replace("\n", "")
    df["Class Code"] = df["Class Code"].str.replace("-", "")
    df["Class Code"] = df["Class Code"].str.replace("  ", " ")
    df = df.reset_index(drop = True)
    list_of_cleaned_df.append(df)
        
```

```python
April2018_Data = pd.concat(list_of_cleaned_df).reset_index(drop=True)
April2018_Data["Date"] = date
April2018_Data = April2018_Data[~April2018_Data["Class Code"].str.startswith('ZZZ')] # drop summary tab
April2018_Data.columns = ['Class Code', 'Item ID', 'Bottle Size', 'Product Name', 'Status', 'Price', 'Cost', 'Total Sales', 'Percent Sales', 'YOY Sales', 'Total Bottle Count','Percent Bottle Count', 'YOY Bottle Count', 'Case Count (9L)', 'Rank', 'Date'] 

April2018_Data['Item ID'] = pd.to_numeric(April2018_Data['Item ID'], errors='coerce')
April2018_Data['Bottle Size'] = pd.to_numeric(April2018_Data['Bottle Size'], errors='coerce')
April2018_Data["Price"] = April2018_Data["Price"].str.replace("  ", "")
April2018_Data["Price"] = April2018_Data["Price"].str.replace("$", "")
April2018_Data['Price'] = pd.to_numeric(April2018_Data['Price'], errors='coerce')
April2018_Data["Cost"] = April2018_Data["Cost"].str.replace("  ", "")
April2018_Data["Cost"] = April2018_Data["Cost"].str.replace("$", "")
April2018_Data['Cost'] = pd.to_numeric(April2018_Data['Cost'], errors='coerce')
April2018_Data['Total Sales'] = pd.to_numeric(April2018_Data['Total Sales'], errors='coerce')
```
