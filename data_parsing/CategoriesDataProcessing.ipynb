```python
import openpyxl
```

```python
wb = openpyxl.load_workbook('data/January-2020-fiscal-period-Sales-Analysis-Report.xlsx', read_only=True)
month = "Jan 2020"
```

```python
def get_sales(ws):
    total_row_index = None
    for row in ws.iter_rows(min_col=3, max_col=3):
        for cell in row:
            if cell.value == "Total": total_row_index = cell

    gross_sales = ws.cell(row=total_row_index.row, column=11)
    return gross_sales.value
```

```python
wb.sheetnames
```

```python
category_totals =[]
for sn in wb.sheetnames:
    ws = wb[sn]
    category_totals.append((ws.title, get_sales(ws)))
```

```python
"""
A - Spirits
C - Liquers
E - Vermouth, Madeira, Sherry, Port (Wine category)
G - Cider, Sangria, mixed, coolers, sake (Wine category)
I - Sparking Wine
K - White Wine
L - Wine offer
M - Rose wine
P - Red wine
R - flavored malt beverage
T - Beer
YGA - Gift sets, spirits
YGC - Gift sets, Liqueurs
YGE - Gift sets, wine
YSA - special order, spirits
YSC - special order, liqueur
YSE - special order, wine
YST - special order, beer
"""
```

```python
category_totals
```

```python
wine = 0
beer = 0
spirits = 0

for code, total in category_totals:
    if code[0] == 'A' or code[0] == 'C' or code == 'YGA' or code == 'YSA':
        if total is not None:
            spirits += total
    elif code[0] == 'E' or code[0] == 'G' or code[0] == 'I' or code[0] == 'K' or code[0] == 'L' or code[0] == 'M' or code[0] == 'P':
        if total is not None:
            wine += total
    elif code == 'YGE' or code == 'YSE':
        if total is not None:
            wine += total
    elif code[0] == 'R' or code[0] == 'T' or code == 'YST':
        if total is not None:
            beer += total
```

```python
# Wine
fortified = 0
fruit_wine = 0
cider = 0
sake = 0
sparkling = 0
white = 0
wine_offer = 0
rose = 0
red = 0
wine_gift_set = 0
wine_spec_order = 0
```

```python
for code, total in category_totals:
    if code[0] == 'E':
        fortified += total
    if code[:2] == 'GF':
        fruit_wine += total
    if code[:2] == 'GL':
        cider += total
    if code[:2] == 'GP':
        sake += total
    if code[:2] in ('IP', 'IH'):
        sparkling += total
    if code[0] == 'K':
        white += total
    if code[0] == 'L':
        wine_offer += total
    if code[0] == 'M':
        rose += total
    if code[0] == 'P':
        red += total
    if code == 'YGE':
        wine_gift_set += total
    if code == 'YSE':
        wine_spec_order += total
```

```python
# Beer
flavored_malt = 0
beer_other = 0
lager = 0
ale = 0
domestic = 0
imported = 0
beer_spec_order = 0
```

```python
for code, total in category_totals:
    if code == 'RCP':
        flavored_malt += total
    if code == 'TCP':
        beer_other += total
    if code[:2] == 'TH':
        lager += total
    if code[:2] == 'TN':
        ale += total
    if code == 'TVB':
        domestic += total
    if code in ['TVD', 'TVF', 'TVH', 'TVN', 'TVP', 'TVU']:
        if total is not None:
            imported += total
    if code == 'YST':
        beer_spec_order += total
```

```python
# Spirits
vodka = 0
gin = 0
rum = 0
tequila = 0
brandy = 0
whiskey = 0
liqueurs = 0
schnapps = 0
premixed = 0
spirits_gift_sets = 0
spirits_spec_order = 0
```

```python
for code, total in category_totals:
    if code[:2] == 'AD':
        vodka += total
    if code[:2] == 'AH':
        gin += total
    if code[:2] == 'AL':
        rum += total
    if code[:2] == 'AP':
        tequila += total
    if code[:2] == 'AT':
        brandy += total
    if code[:2] == 'AW':
        whiskey += total
    if code[:2] == 'CH':
        liqueurs += total
    if code[:2] == 'CP':
        schnapps += total
    if code[:2] == 'CW':
        premixed += total
    if code == 'YGA':
        spirits_gift_sets += total
    if code == 'YSA':
        spirits_spec_order += total
```
