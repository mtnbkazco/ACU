```python
#January2019
January2019 = pd.read_excel("./Desktop/January2019.xlsx", sheet_name = None)
```

```python
list_of_dfs = []
for i in January2019.keys():
    list_of_dfs.append(January2019[i])

date = list_of_dfs[0].columns[0].strftime("%Y-%d-%m")
date
```

```python
for df in list_of_dfs:
    df = df.fillna(value=np.nan)

list_of_cleaned_df = []
for df in list_of_dfs:
            df = df.fillna(value=np.nan)
            class_code = df.iloc[:,1:2].values.tolist()
            class_code = [item for sublist in class_code for item in sublist]
            class_code = list(set(class_code))
            class_code = [i for i in class_code if len(str(i)) > 9][0]
            class_code = class_code.replace("\n", "")
            class_code = class_code.replace("-", "")
            class_code = class_code.replace("  ", " ")
            
            df = df.dropna(thresh=13)
            column_s = df.dropna(thresh=15).values.tolist()[0]
            df.columns = column_s
            df = df[1:]
            df = df.loc[:, df.columns.notnull()]
            df["Class Code"] = class_code
            df = df.reset_index(drop = True)

            list_of_cleaned_df.append(df)      
```

```python
January2019_Data = pd.concat(list_of_cleaned_df).reset_index(drop=True)
January2019_Data
```

```python
January2019_Data["Date"] = date
January2019_Data = January2019_Data[~January2019_Data["Class Code"].str.startswith('ZZZ')] # drop summary tab
January2019_Data.columns = ['Class Code', 'Item ID', 'Bottle Size', 'Product Name', 'Status', 'Price', 'Cost', 'Total Sales', 'Percent Sales', 'YOY Sales', 'Total Bottle Count','Percent Bottle Count', 'YOY Bottle Count', 'Case Count (9L)', 'Rank', 'Date'] 

January2019_Data['Item ID'] = pd.to_numeric(January2019_Data['Item ID'], errors='coerce')
January2019_Data['Bottle Size'] = pd.to_numeric(January2019_Data['Bottle Size'], errors='coerce')
```

```python
#January2019_Data["Price"] = January2019_Data["Price"].str.replace("  ", "")
#January2019_Data["Price"] = January2019_Data["Price"].str.replace("$", "")
January2019_Data['Price'] = pd.to_numeric(January2019_Data['Price'], errors='coerce')
#January2019_Data["Cost"] = January2019_Data["Cost"].str.replace("  ", "")
#January2019_Data["Cost"] = January2019_Data["Cost"].str.replace("$", "")
January2019_Data['Cost'] = pd.to_numeric(January2019_Data['Cost'], errors='coerce')
January2019_Data['Total Sales'] = pd.to_numeric(January2019_Data['Total Sales'], errors='coerce')
```
