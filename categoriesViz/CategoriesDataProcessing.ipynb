```python
import openpyxl
```

```python
wb = openpyxl.load_workbook('data/Sales_Analysis_Report_P3_FY21.xlsx', read_only=True)
```

```python
def get_sales(ws):
    total_row_index = None
    for row in ws.iter_rows(min_col=3, max_col=3):
        for cell in row:
            if cell.value == "Total": total_row_index = cell

    gross_sales = ws.cell(row=total_row_index.row, column=10)
    return gross_sales.value
```

```python
wb.sheetnames
```

```python
category_totals =[]
for sn in wb.sheetnames:
    ws = wb[sn]
    category_totals.append((ws.title, get_sales(ws)))
```

```python
"""
A - Spirits
C - Liquers
E - Vermouth, Madeira, Sherry, Port (Wine category)
G - Cider, Sangria, mixed, coolers, sake (Wine category)
I - Sparking Wine
K - White Wine
L - Wine offer
M - Rose wine
P - Red wine
R - flavored malt beverage
T - Beer
YGA - Gift sets, spirits
YGC - Gift sets, Liqueurs
YGE - Gift sets, wine
YSA - special order, spirits
YSC - special order, liqueur
YSE - special order, wine
YST - special order, beer
"""
```

```python
category_totals
```

```python
wine = 0
beer = 0
spirits = 0

for code, total in category_totals:
    if code[0] == 'A' or code[0] == 'C' or code == 'YGA' or code == 'YSA':
        if total is not None:
            spirits += total
    elif code[0] == 'E' or code[0] == 'G' or code[0] == 'I' or code[0] == 'K' or code[0] == 'L' or code[0] == 'M' or code[0] == 'P':
        if total is not None:
            wine += total
    elif code == 'YGE' or code == 'YSE':
        if total is not None:
            wine += total
    elif code[0] == 'R' or code[0] == 'T' or code == 'YST':
        if total is not None:
            beer += total
```
